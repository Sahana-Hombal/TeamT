```python
## 1(d) ##

## Used to import XML file
import xml.etree.ElementTree as ET
import csv

## To load XML data
tree = ET.parse('C:/Users/sujit/OneDrive/Desktop/Database and Analytics Programming/web-3Dayforecast.xml')
root = tree.getroot()

## Define CSV file and column names
csv_file = 'weather_forecast.csv'
fields = ['Id', 'location', 'day_num', 'wind_speed', 'wind_dir']

##  CSV file for writing
with open(csv_file, mode='w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=fields)
    
    ## Column names to CSV file
    writer.writeheader()
    
    ## Iterate through XML data and extract required information
    for forecast in root.findall('.//forecast'):
        id = forecast.find('Id').text
        location = forecast.find('location').text
        day_num = forecast.find('day_num').text
        wind_speed = forecast.find('wind_speed').text
        wind_dir = forecast.find('wind_dir').text
        
        ## Extracted information to CSV file
        writer.writerow({'Id': id, 'location': location, 'day_num': day_num, 'wind_speed': wind_speed, 'wind_dir': wind_dir})

print("CSV file has been created successfully.")

import xml.etree.ElementTree as ET
import csv

# Parse the XML data
tree = ET.parse('your_xml_file.xml')
root = tree.getroot()

# Define the CSV file path
csv_file_path = 'weather_forecast.csv'

# Open the CSV file in write mode
with open(csv_file_path, mode='w', newline='') as file:
    # Define the CSV writer
    writer = csv.writer(file)
    
    # Write the column names
    writer.writerow(['Id', 'location', 'day_num', 'wind_speed', 'wind_dir'])
    
    # Iterate through each station
    for station in root.findall('.//station'):
        station_id = station.find('id').text
        location = station.find('location').text
        
        # Iterate through each day within the station
        for day in station.findall('.//day'):
            day_num = day.find('day_num').text
            wind_speed = day.find('wind_speed').text.strip()
            wind_dir = day.find('wind_dir').text
            
            # Write the data to the CSV file
            writer.writerow([station_id, location, day_num, wind_speed, wind_dir])

print("CSV file has been created successfully.")
```

```python
## 3(a) ##

import re
## Function using regular expression
def highlight_two_letter_words(text):
    highlighted_text = re.sub(r'\b[a-zA-Z]{2}\b', r'{\g<0>}', text) ## \b denotes the word boundaries
    return highlighted_text

## Example
text = "GPT don't chat\nTo me or yourself\nUnless it is noon\nOn the Moon."
highlighted_text = highlight_two_letter_words(text)
print(highlighted_text)

```

```python
## 3(b) ##

import re
## Function using regular expression

def highlight_first_three_characters(text):
    highlighted_text = re.sub(r'^(.{0,3})', r'{\1}', text, flags=re.MULTILINE)
    return highlighted_text

## Example
text = """GPT don't chat
To me or yourself
Unless it is noon
On the Moon."""
highlighted_text = highlight_first_three_characters(text)
print(highlighted_text)
```
